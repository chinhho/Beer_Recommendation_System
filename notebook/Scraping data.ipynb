```python
def scraping(beer): #scraping a beer
    beers={}
    beers['Beers']= beer.find('a').text
    beers['Brewery']= beer.find('a').findNext('a').text
    beers['abv'] = beer.find('span').text
    beers['Url'] = beer.find('a')['href']
    return beers
```

```python
ex=[scraping(beer) for beer in first_page]
```

```python
ex
```

```python
def click(link): #lay ten beer
    data=[]
    html=requests.get(link)
    soup=BeautifulSoup(html.content, 'html.parser')
    first_page=soup.find_all('tr')
    first_page = list(filter(lambda row: not row.find('td', {'colspan':"6"}), first_page))
    first_page = list(filter(lambda row: not row.find('td', {'colspan':"5"}), first_page))
    first_page = list(filter(lambda row: not row.find('td', {'bgcolor':"#F0F0F0"}), first_page))
    data.extend([scraping(beer) for beer in first_page])
    return data
```

```python
all_beers=[]
for link in urls:
    all_beers.extend(click(link))
```

```python
def scraping_raters(r):
    raters={}
    raters['Id']=r['ba-user']
    raters['Rater']=r.find('b').text
    raters['From']=r.find('b').next_sibling
    raters['Beer Id']=r.find('span', {'class':'muted'}).findNext('span', {'class':'muted'}).findNext('span', {'class':'muted'}).find('a')['href'] 
    raters['Rating']=r.find('span', {'class':'BAscore_norm'}).text
    return raters
```

```python
all_raters=[]
def scraping_all_raters(link):
    html=requests.get(link)
    soup=BeautifulSoup(html.content, 'html.parser')
    first25=soup.find_all('div', {'id':'rating_fullview_container'})
    all_raters.extend(scraping_raters(r) for r in first25)
    return all_raters
```
